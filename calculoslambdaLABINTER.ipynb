# LED VERDE

```python
a=loadtxt('datosEspectros/verde.dat')
pixel=a[:,0]
value=a[:,1]
lamb=791.3866346421-(pixel*0.544351006)#Formula resultante de la calibracion con la lampara de hidrogeno
plt.suptitle('En pixeles')
plot(pixel,value)
xlabel('Pixel')
ylabel('#Electrones')
show()
plt.suptitle('En longitudes de onda')
plot(lamb,value)
xlabel('Lambda[nm]')
ylabel('#Electrones')
show()
```

```python
expe=loadtxt('datosMonocromador/LED_verde1.dat')
posini=4.7 #en milimetros
paso=expe[:,0]
vale=expe[:,1]
plt.suptitle('Datos experimentales')
plot(paso, vale)
xlabel('Pasos en el tornillo')
ylabel('Voltaje')#Voltaje medido en el osciloscopio
show()
```
